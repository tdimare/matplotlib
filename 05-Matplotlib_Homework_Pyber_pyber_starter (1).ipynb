```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

#Combine the data into a single dataset
city_ride_data = pd.merge(ride_data, city_data, how="left", on=["city","city"])

#Display the data table for preview
city_ride_data.head(10000)
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_city = city_ride_data[city_ride_data["type"]=="Urban"]
rural_city = city_ride_data[city_ride_data["type"]=="Rural"]
suburban_city = city_ride_data[city_ride_data["type"]=="Suburban"]


urban_ride_count = urban_city.groupby(["city"]).count()["ride_id"]
urban_average_fare = urban_city.groupby(["city"]).mean()["fare"]
urban_driver_count = urban_city.groupby(["city"]).mean()["driver_count"]

rural_ride_count = rural_city.groupby(["city"]).count()["ride_id"]
rural_average_fare = rural_city.groupby(["city"]).mean()["fare"]
rural_driver_count = rural_city.groupby(["city"]).mean()["driver_count"]

suburban_ride_count = suburban_city.groupby(["city"]).count()["ride_id"]
suburban_average_fare = suburban_city.groupby(["city"]).mean()["fare"]
suburban_driver_count = suburban_city.groupby(["city"]).mean()["driver_count"]

# Build the scatter plots for each city types
#urban city
plt.scatter (urban_ride_count,
          urban_average_fare,
          s=10*urban_driver_count, c="coral",
          edgecolor="green",linewidths=1,marker="o",
          alpha=0.8, label="Urban")

#rural city
plt.scatter (rural_ride_count,
          rural_average_fare,
          s=10*rural_driver_count, c="blue",
          edgecolor="green",linewidths=1,marker="o",
          alpha=0.8, label="Rural")

#suburban city
plt.scatter (suburban_ride_count,
          suburban_average_fare,
          s=10*suburban_driver_count, c="yellow",
          edgecolor="green",linewidths=1,marker="o",
          alpha=0.8, label="Suburban")

# Incorporate the other graph properties
plt.grid(True)
plt.title("Pyber Ride Sharing Data")
plt.ylabel("Averaeg Fare Amount")
plt.xlabel("Average Ride Amount")

# Create a legend
lgnd = plt.legend(fontsize="medium", mode="Expanded",
                 numpoints=1, scatterpoints=1,
                 loc="best", title="City Types",
                 labelspacing=0.6)
lgnd.legendHandles[0]._sizes = [30]
lgnd.legendHandles[1]._sizes = [30]
lgnd.legendHandles[2]._sizes = [30]

# Incorporate a text label regarding circle size
plt.text(45, 40, "Note: Circle size correlates to number of drivers.", size=20, rotation=0.,
         ha="left", va="baseline")
plt.show()
# Save Figure
plt.savefig("scatterchart.jpg")

```

## Total Fares by City Type

```python
# Calculate Type Percents
fares_percent = 100 * city_ride_data.groupby(["type"]).sum()["fare"]

# Build Pie Chart
plt.title("Percent of Total Fares by Type of City")
plt.pie(fares_percent, labels=["Rural", "Suburban", "Urban"], colors=["yellow", "blue", "coral"], explode=[0.0,0,0.2], autopct='%1.1f%%', shadow=True, startangle=100)

# Save Figure
plt.savefig("faresbycity.jpg")
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_percent = 100 * city_ride_data.groupby(["type"]).count()["ride_id"]

# Build Pie Chart
plt.title("Percent of Total Rides by Type of City")
plt.pie(ride_percent, labels=["Rural", "Suburban", "Urban"], colors=["yellow", "blue", "coral"], explode=[0.0,0,0.2], autopct='%1.1f%%', shadow=True, startangle=100)

# Save Figure
plt.savefig("ridesbycity.jpg")
# Show Figure
plt.show()
```

```python
# Calculate Driver Percents
driver_percent = 100 * city_ride_data.groupby(["type"]).sum()["driver_count"]

# Build Pie Chart
plt.title("Percent of Total Rides by Type of City")
plt.pie(driver_percent, labels=["Rural", "Suburban", "Urban"], colors=["yellow", "blue", "coral"], explode=[0.0,0,0.2], autopct='%1.1f%%', shadow=True, startangle=100)

# Save Figure
plt.savefig("ridesbycity.jpg")
# Show Figure
plt.show()
```
